```python
# importing pandas as pd
import pandas as pd
import numpy as np  
# Create the Timestamp object


julian_dates = []
year_value = np.array([2010, 2011])
month_value = np.arange(1,13)

month_31days = np.array([1,3,5,7,8,10,12])
month_30days = np.array([4,6,9,11])
month_28days = np.array([2])

for year in year_value:
    for month in month_value:
        if month in month_31days:
            day_values = np.arange(1,32)
        elif month in month_30days:
            day_values = np.arange(1, 31)
        elif month in month_28days:
            day_values = np.arange(1, 29)
        for day in day_values:
            ts = pd.Timestamp(year = year,  month = month, day = day)
            julian_dates.append(ts.to_julian_date())

julian_dates = np.array(julian_dates)
print(julian_dates)

#we get the julian dates of each day between 2010/1/1 to 2011/12/31
```
